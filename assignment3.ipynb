```python
#                                          ANSWER1
#A "for" loop is used to run a block of code for a certain number of times.
#example:

print("DISPLAY NUMBERS 1 TO 10")
for i in range(1,11):
    print(i , end = " ")
    
#A "While" Loop is used to repeat a specific block of code an unknown number of times, until a condition is met..

print("")
print("DISPLAY NUMBERS 1 TO 10")
n=1
while n<=10:
    print(n , end = " ")
    n = n+1
```

```python
#                                         ANSWER 2
#BY FOR LOOP
sum=0
product=1
for i in range(1,11):
    sum = sum+i
    product = product*i
print(" SUM OF FIRST 10 NATURAL NUMBER = ",sum)
print(" PRODUCT OF FIRST 10 NATURAL NUMBER = ",product)
#BY WHILE LOOP
sum1=0
product1=1
n=1
while n<=10:
    sum1 = sum1 + n
    product1 = product1 * n
    n = n + 1
print(" SUM OF FIRST 10 NATURAL NUMBER = ",sum1)
print(" PRODUCT OF FIRST 10 NATURAL NUMBER = ",product1)
```
